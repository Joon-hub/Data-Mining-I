#### 8.1 Analyzing the Shopping Basket Data Set

##### 8.1.1 The Shopping Basket data set is provided as an Excel file in ILIAS. Load the data set with the read_excel function.

````python
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel('ShoppingBaskets.xls')

# Drop the 'BasketNo' column from the DataFrame
df = df.drop('BasketNo', axis=1)

# Display the first few rows of the DataFrame
display(df.head(10))
````

##### 8.1.2 Mine frequent item sets from the data set using the frequent_itemsets function (support =0.2). Which items are usually bought together with the laptop, the netbook and the printer?

````python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support=0.2, use_colnames= True)
frequent_itemsets 
````

````python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.7)
rules
````

````python
laptop_rules = rules[rules['antecedents'].astype(str).str.contains('ThinkPad X220')]
printer_rules = rules[rules['antecedents'].astype(str).str.contains('HP Laserjet P2055')]
Netbook_rules = rules[rules['antecedents'].astype(str).str.contains('Asus EeePC')]
````

#### 8.2 Finding Frequent Pattern in the Adult Data Set

##### 8.2.1 Import the Adult-tweaked data set and print a description of the dataset using the describe function. The Adult-tweaked data set is provided on ILIAS as an ARFF file.

````python
from scipy.io import arff
adult_arff_data, adult_arff_meta = arff.loadarff(open('adult-dataset-tweaked.arff', 'r'))
adult = pd.DataFrame(adult_arff_data)
adult = adult.applymap(lambda x: x.decode('utf8') if hasattr(x, 'decode') else x)
adult.head()
````

##### 8.2.2 Prepare the data set for Frequent Pattern Mining by: 1) reducing the size of the data set to 5000 examples using sampling; 2) using a subset of attributes: age, workclass, education, occupation, race, sex, hours-per-week, native-country and class; 3) discretizing the attributes age and hours-per-week into three user defined ranges (think about ranges that could make sense for the attributes). How many attributes does the resulting data set have?

````python
# 1) reducing the size of the data set to 5000 examples using sampling;
sampled_df = adult.sample(n = 5000, random_state=42, axis = 0)

# 2) using a subset of attributes: age, workclass, education, occupation, race, sex, hours-per-week, native-country and class;

subset_df = sampled_df[['age','workclass','education','occupation','race','sex','hours-per-week','native-country','class']]
subset_df.head()

# 3) discretizing the attributes age and hours-per-week into three user defined ranges (think about ranges that could make sense for the attributes). 
# How many attributes does the resulting data set have?

import warnings
warnings.filterwarnings("ignore")

# from sklearn.preprocessing import KBinsDiscretizer
# import matplotlib.pyplot as plt
# encoder = KBinsDiscretizer()

# plt.figure(figsize = (10,3))

# for i in range(1,3):
#     axs = plt.subplot(1,2,i)
#     if i ==1:
#         axs.hist(subset_df[['age']])
#     elif i == 2:
#         axs.hist(subset_df[['hours-per-week']])
# plt.tight_layout()
# plt.show()


subset_df['age'] = pd.cut(subset_df['age'],[0,20,65,100],labels= ['low','middle','high'])
subset_df['hours-per-week'] = pd.cut(subset_df['hours-per-week'],[0,20,40,100],labels= ['part-time','full-time','over-time'])
display(subset_df.head())

# one hot encoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse= False)
oneHot_data = encoder.fit_transform(subset_df)
names = encoder.get_feature_names_out(subset_df.columns)
df = pd.DataFrame(oneHot_data,columns = names)
display(df)
````

##### 8.2.3 Find frequent item sets that have a support above 0.2. What can you learn from these item sets about the people who earn less than 50K a year?

````python
from mlxtend.frequent_patterns import apriori

#replace column numbers with their names
frequent_patterns = apriori(df[df['class_<=50K'] == 1], min_support=0.2, use_colnames= True)
display(frequent_patterns)
````

##### 8.2.4 Given the large number of examples and the low min-support threshold, the number of frequent item sets containing the education attribute is surprisingly low. Moreover, only one value for this attribute is present in the resulting frequent item sets. Why is this the case? How could you aggregate the data to change this without losing too much information? Also look at the native-country attribute and think of a possible aggregation for it.

````python
# binning the education column
subset_df['education'] = ['Basic education' if i in ['12th','HS-grad','11th','10th','7th-8th','9th','5th-6th','1st-4th','Preschool'] else i for i in subset_df['education']]
subset_df['education'] = ['vocational education' if i in ['Assoc-voc','Assoc-acdm','Prof-school','Some-college'] else i for i in subset_df['education']]
subset_df.loc[subset_df['native-country'] != 'United-States', 'native-country'] = 'Non-US'
display(subset_df.head())

# one hot encoder
encoder = OneHotEncoder(sparse= False)
oneHot_data = encoder.fit_transform(subset_df)
names = encoder.get_feature_names_out(subset_df.columns)
subset_df_encoded = pd.DataFrame(oneHot_data,columns = names)
display(subset_df_encoded)
````

##### 8.2.5 Restrict the frequent item sets to the ones containing “class = >50K” and lower the support so that a decent number of item sets is discovered. What can you learn from these item sets about the people who earn more than 50K a year?


````python
#frequent item sets containing 'class_>50K

from mlxtend.frequent_patterns import apriori
column_mapping = {index:col_names for index,col_names in enumerate(subset_df_encoded.columns)} 
frequent_itemsets = apriori(subset_df_encoded, min_support=0.2)
frequent_itemsets['itemsets'] = frequent_itemsets['itemsets'].apply(lambda x: [column_mapping[item] for item in x])
display(frequent_itemsets[frequent_itemsets['itemsets'].astype(str).str.contains('class_>50K')])
````

Support, confidence, and lift are common metrics used in association rule mining to evaluate the quality and significance of discovered rules. 

1. **Low support and confidence with lift close to 1**: 
   - **Support**: Support measures the frequency of occurrence of an itemset in the dataset. Low support indicates that the rule is not frequently applicable to the dataset.
   - **Confidence**: Confidence measures the reliability of the rule. Low confidence indicates that the rule has a low predictive power.
   - **Lift**: Lift measures how much more often the antecedent and consequent of a rule occur together compared to what would be expected if they were independent. A lift close to 1 indicates that there is no significant association between the antecedent and consequent beyond what would be expected by chance.
   - Scenario 1 likely indicates an uninteresting rule because both support and confidence are low, meaning the rule is infrequent and not reliable. Additionally, the lift close to 1 suggests that there is no significant association between the antecedent and consequent beyond chance.

2. **High support and low confidence with lift close to 1**:
   - **Support**: High support indicates that the itemset occurs frequently in the dataset.
   - **Confidence**: Low confidence indicates that although the rule is frequently applicable, it doesn't always hold true.
   - **Lift**: Lift close to 1 suggests that there is no significant association between the antecedent and consequent beyond what would be expected by chance.
   - Scenario 2 may indicate an interesting rule because it has high support, meaning the rule is applicable to a significant portion of the dataset. However, the low confidence suggests that the rule may not always hold true. The lift close to 1 indicates that the association between the antecedent and consequent is not significant beyond what would be expected by chance. 

In summary, while scenario 1 indicates an uninteresting rule due to low support, low confidence, and lift close to 1, scenario 2 may indicate an interesting rule due to high support, low confidence, and lift close to 1. However, further analysis and domain knowledge are necessary to determine the actual significance and usefulness of the discovered rules.

Apriori and FP-Growth are both algorithms used for frequent itemset mining in association rule learning, but they use different approaches to achieve this goal. Here's a comparison of the two algorithms:

1. **Apriori Algorithm**:
   - Apriori is a classical algorithm for frequent itemset generation.
   - It works by generating candidate itemsets of length \( k \) from itemsets of length \( k-1 \) using a level-wise approach.
   - It scans the database multiple times to find frequent itemsets.
   - It uses a candidate generation-and-test approach, which can be computationally expensive, especially for large datasets and itemsets.

2. **FP-Growth Algorithm**:
   - FP-Growth (Frequent Pattern Growth) is a more modern and efficient algorithm compared to Apriori.
   - It constructs a compact data structure called an FP-tree (Frequent Pattern tree) to represent the dataset.
   - FP-Growth mines frequent itemsets directly from the FP-tree without the need for candidate generation.
   - It requires only two passes over the dataset, making it much faster and more memory-efficient than Apriori, especially for large datasets.

Example:

Let's say we have a transaction dataset containing the following transactions:

Transaction 1: {bread, milk, eggs}
Transaction 2: {bread, butter}
Transaction 3: {milk, butter}
Transaction 4: {bread, milk}

Using the Apriori algorithm:

- Suppose we set a minimum support threshold of 50%.
- In the first pass, we identify frequent single items: {bread}, {milk}, {butter}.
- In the second pass, we generate candidate pairs and prune those below the minimum support threshold: {(bread, milk)}, {(bread, butter)}, {(milk, butter)}.
- We find that only {(bread, milk)} meets the minimum support threshold.

Using the FP-Growth algorithm:

- We construct an FP-tree from the transactions.
- The FP-tree for the given transactions looks like this:
  ```
       null (root)
       /   |   \
    bread milk butter
     /     \
  milk   butter
  ```
- From the FP-tree, we can directly identify frequent itemsets without generating candidates.
- Frequent itemsets found: {bread}, {milk}, {butter}, {(bread, milk)}.

While both algorithms produce the same result, FP-Growth typically outperforms Apriori in terms of efficiency, especially on large datasets, because it doesn't require candidate generation and multiple passes over the data.
